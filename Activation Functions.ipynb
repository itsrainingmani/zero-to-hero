# Activation Functions

Inputs are like $\ w_n x_n $ all going into a cell body. The cell body sums up the inputs $$\sum_{i} w_i x_i + b$$ where b is the bias (innate trigger happiness). We take this through an activation function like sigmoid or tanh (input gets squashed on the y-axis)

S the output of the neuron is - $$f(\sum_{i} w_i x_i + b)$$ where f is the activation fn

## tanh broken down

$ tanh(x) = \dfrac{e^2x - 1}{e^2x + 1}$

```python
plt.plot(np.arange(-5, 5, 0.2), np.tanh(np.arange(-5, 5, 0.2)))
plt.grid()
```

## Sigmoid Activation Function

$ y(x) = \dfrac{1}{1 + e^-x}$

### Drawbacks

- Non-zero centered
- Model learning rate is slow
- Creates the vanishing gradient problem

```python
sigmoid = 1 / (1 + np.exp(-np.arange(-5, 5, 0.2)))
plt.plot(np.arange(-5, 5, 0.2), sigmoid)
plt.grid()
```

```python
relu = lambda x: 0 if x <= 0 else x

plt.plot(np.arange(-5, 5, 0.2), [relu(x) for x in np.arange(-5, 5, 0.2)])
plt.grid()
```
